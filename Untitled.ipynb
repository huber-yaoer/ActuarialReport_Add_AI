### 准备中国地图

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
imageFile_china = './china.png'
img_china = cv2.imread(imageFile_china)
# ret,thresh3 = cv2.threshold(img_china,127,255,cv2.THRESH_BINARY) #二值化
# cv2.imwrite("./china2thresh.png",thresh3)
plt.imshow(img_china, cmap = "gray")
plt.xticks([]), plt.yticks([])  # 隐藏坐标轴
plt.show()
```

### 用画图软件加入黑色散点（潜在会员数），另存为china_people.png，再执行程序：

注意：潜在会员数的密度，总数，分布可自由控制，可参考QQ在线人数：https://www.gznotes.com/manual/qqOnline/

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
imageFile_china = './china2thresh.png'
imageFile_china_people = './china_people.png'
img_china = cv2.imread(imageFile_china)
img_china_people = cv2.imread(imageFile_china_people)
img_people = img_china-img_china_people
r_img_china = cv2.split(img_china)[0]
r_img_people = cv2.split(img_people)[2]
cv2.imwrite("./people.png",r_img_people)

print(r_img_china.shape, r_img_china.shape[0]*r_img_china.shape[1])
print(np.unique(r_img_china, return_counts=True))
#print(np.unique(img_china_people, return_counts=True))
print(np.unique(r_img_people, return_counts=True))

print('中国大陆面积（像素数）：', np.unique(r_img_china, return_counts=True)[1][1])
print('小镇中国区潜在会员数（像素数）：', np.unique(r_img_people, return_counts=True)[1][1])


#  用于存储所有弹框的图片集合
psw=[]
psw.append(("china",img_china))
psw.append(("china+people",img_china_people))
psw.append(("people",img_people))


# 获取个数
plot_number=len(psw)
# 设置每列显示的窗体个数
cols=3
# 行数自动推算
rows=plot_number/cols+1
# 打印所有的图片
for index in range(plot_number):
    plt.subplot(rows, cols, index + 1)
    plt.imshow(psw[index][1], cmap = "gray")
    plt.title(psw[index][0], fontsize=8)
    plt.xticks([]), plt.yticks([])  # 隐藏坐标轴

plt.show()
```

### 用people图构造KMeans训练集

```python
print(r_img_people.shape)
X = np.argwhere(r_img_people>0)
print(X.shape)
X[0:10] #所有人（白点）的坐标
```

### 训练KMeans模型

```python
from sklearn.cluster import KMeans
import copy
k = 72 #分为k个簇
y_KMeans = KMeans(n_clusters=k, random_state=9)
y_pred = y_KMeans.fit_predict(X)
print(y_pred)

src_img_china = copy.deepcopy(img_china)
r_src_img_china = cv2.split(src_img_china)[0]
#根据索引赋值
for i in range(len(X)):
    index = X[i]
    src_img_china[index[0]][index[1]] = [y_pred[i]*(255/k), (255/3 + y_pred[i]*(255/k))%255, 255*2/3+(y_pred[i]*(255/k))]
```

### 可视化结果

```python
cv2.imwrite("./result.png",cv2.cvtColor(src_img_china, cv2.COLOR_RGB2BGR))
plt.imshow(src_img_china)
plt.xticks([]), plt.yticks([])  # 隐藏坐标轴
plt.show()

plt.scatter(X[:, 1], X[:, 0], c=y_pred)
plt.xticks([]), plt.yticks([])  # 隐藏坐标轴
ax = plt.gca() 
ax.invert_yaxis()  #y轴反向
plt.show()
```

### 总结输出：

```python
num_pix_china = np.unique(r_img_china, return_counts=True)[1][1]
print('中国大陆面积（像素数）：', num_pix_china)
num_all_people = np.unique(r_img_people, return_counts=True)[1][1]
print('小镇中国区潜在会员数（像素数）：', num_all_people)
print('每个簇的中心点坐标（像素坐标）：', y_KMeans.cluster_centers_)
print('每个簇的潜在会员数（簇内像素数）：',np.unique(y_pred, return_counts=True))

#每个簇内到簇中心距离列表：
# print(X.shape)
# print(y_pred.shape)
dic_result_xy = {}
dic_result = {}
for i in range(y_pred.shape[0]):
    if y_pred[i] not in dic_result:
        dic_result_xy[y_pred[i]]=[]
        dic_result[y_pred[i]]=[]
    dic_result_xy[y_pred[i]].append(X[i])
#     print('jjjjjjjjj',i,X[i][0]-y_KMeans.cluster_centers_[y_pred[i]][0])
#     print('kkkkkkkkk',i,X[i][1]-y_KMeans.cluster_centers_[y_pred[i]][1])
    dic_result[y_pred[i]].append( np.sqrt( np.square(X[i][0]-y_KMeans.cluster_centers_[y_pred[i]][0])\
                                                  + np.square(X[i][1]-y_KMeans.cluster_centers_[y_pred[i]][1])))
    if i%5000 == 0:
        print('计算到第{}个循环'.format(i))
# print(dic_result_xy[0][:10])
# print(dic_result[0])


#每个簇内到簇中心平均距离：
tmp = []
for i in range(k):
    tmp.append(np.mean(dic_result[i]))
print('每个簇簇内所有点到簇中心平均距离（像素距离）：')
print(tmp)
   
tmp = []
for i in range(k):
    tmp.append(np.median(dic_result[i]))
print('每个簇簇内所有点到簇中心中位数距离（像素距离）：')
print(tmp)

tmp = []
for i in range(k):
    tmp.append(max(dic_result[i]))
print('每个簇簇内所有点到簇中心最远距离（像素距离）：')
print(tmp)
print(np.mean(tmp))

actual_china = 9083674#纯陆地9,083,674平方千米
pix2km = np.sqrt(actual_china/num_pix_china)
print('最终:', np.mean(tmp)*pix2km,'km')
```
